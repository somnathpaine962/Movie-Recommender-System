Now we have to modify the columns and create a new column using existings
for thta first i have convert the datatype from string to list.

```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)  #this will convert string to list
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#we will keep only top 3 actors from the cast column
def convert2(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
           l.append(i['name'])
           counter +=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head(2)
```

from crew we just need the director name

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
     
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies
```

for removing the space between names

```python
def collapse(L):
    l1=[]
    for i in L:
        l1.append(i.replace(" ",""))
    return l1
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
```

now convert the overview column into list

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['id','title','tag']]
```

```python
new_df.head()
```

here we will take similar words as same like love ,loving,loved etc using the following library.this it apply only on the string not in list

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y) # convert back to string
```

```python
#join in space 
new_df['tag']=new_df['tag'].apply(lambda x : " ".join(x))
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
#make all words to lower case in tag
new_df['tag']=new_df['tag'].apply(lambda x: x.lower())
```

Now we have to do text vectorization.
technique to convert text data to vector is "bags of words".

```python
#to change word to vectors
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tag']).toarray()
vector
```

```python
vector.shape
```

Now we will calculate the distance among the vectors.
we will cosine distance between vector i.e the angel between the vectors.
we use cosine distance because euclideian distance is not suitable for heigher dimensional space.

distance is inversely proposnal to similarity.

there is a function in sklearn for cosine distance calculation.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity[1]
```

```python
new_df[new_df['title']=='Batman Begins'].index[0]
```

```python
sorted(list(enumerate(similarity[new_df[new_df['title']=='Batman Begins'].index[0]])),reverse=True,key=lambda x :x[1])
```

```python
def recommend(movie):
       index=new_df[new_df['title']==movie].index[0]
       distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x :x[1])
       for i  in distances[1:6]:
              print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
